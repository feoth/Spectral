```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mat_mat(n):
    a = np.random.randn(n, n)
    b = np.random.randn(n, n*n)
    res = %timeit -o -q a.dot(b)
    print(n, res.average)
    return res.average
```

```python
N = 2**np.arange(9)
timings = [mat_mat(n) for n in N]
```

```python
plt.plot(N, timings)
plt.plot(N, 1.2e-11*N**4)
plt.ylim(1e-7, None)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('n')
plt.ylabel('timing (s)')
```

```python
def fourier(n):
    a = np.random.randn(n*n, n)
    res = %timeit -o -q np.fft.fft(a)
    print(n, res.average)
    return res.average
```

```python
timings_f = [fourier(n) for n in N]
```

```python
def sum(n):
    a = np.random.randn(n, n*n)
    b = np.random.randn(n, n*n)
    res = %timeit -o -q a+b
    print(n, res.average)
    return res.average
```

```python
timings_s = [sum(n) for n in N]
```

```python
plt.plot(N, timings, label='mat')
plt.plot(N, timings_f, label='fourier')
plt.plot(N, timings_s, label='sum')
plt.plot(N, 1.2e-11*N**4)
plt.legend()
plt.ylim(1e-7, None)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('n')
plt.ylabel('timing (s)')
```
