```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
plt.rcParams['animation.html'] = 'html5'
import numpy as np
import scipy as sp
import scipy.integrate
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from spectral import *
np.set_printoptions(linewidth=120)
```

```python
Lx, Hy, Hz = 1000, 2, 2 # millimeters
Dx = 10 # domain size in millimeters
mesh_x = Multidomain(Legendre(12), np.linspace(0, Lx, int(round(Lx/Dx)) + 1), periodic=False)
mesh_y = Legendre(6, (-Hy/2, Hy/2))
mesh_z = Legendre(6, (-Hz/2, Hz/2))
mesh = mesh_x*mesh_y*mesh_z

rho = 1.06; young = 3.7; nu = 0.34; l = -18.9; m = -13.3; n = -10.0 # PS
nlc = 1
l *= nlc; m *= nlc; n *= nlc
#rho = 1.18; young = 5; nu = 0.34; l = -18.9; m = -13.3; n = -10.0 # PMMA (?)
#rho = 8.4; young = 110; nu = 0.34 # brass
lam = young*nu/(1 + nu)/(1 - 2*nu)
mu = young/2/(1 + nu)
```

```python
def cauchy_green(u):
    deform = u.grad()
    return 1/2*(deform + deform.T + deform@deform.T)

def kin_energy(v):
    return rho*(v@v).int()/2

def pot_energy(u):
    CG = cauchy_green(u)
    I1 = CG.trace()
    I2 = 1/2*(CG.trace()**2 - (CG@CG.T).trace())
    I3 = CG.det()
    pot = (lam/2 + mu)*I1**2 - 2*mu*I2 + (l + 2*m)/3*I1**3 - 2*m*I1*I2 + n*I3
    return pot.int()

def energy(u, v):
    return kin_energy(v) + pot_energy(u)
```

```python
def compress(u, v):
    return np.stack((u.func, v.func)).ravel()

def decompress(y):
    func_u, func_v = y.reshape(2, 3, *mesh.shape)
    u = TensorField(mesh, func_u)
    v = TensorField(mesh, func_v)
    return u, v

def derivative(t, y):
    u, v = decompress(y)
    du_dt = v
    
    CG = cauchy_green(u)
    I1 = CG.trace()
    I2 = 1/2*(CG.trace()**2 - (CG@CG.T).trace())
    dI1 = np.eye(3)
    dI2 = dI1*CG.trace() - CG
    dI3 = CG.cofactor_matrix()
    
    PK = ((lam + 2*mu)*I1*dI1 - 2*mu*dI2 + (l + 2*m)*I1**2*dI1 
          - 2*m*(dI1*I2 + I1*dI2) + n*dI3)@(np.eye(3) + u.grad())
    
    bval = (0,0), (0, 0), (0, 0)
    F = PK.div(bval)
    
    F = F.match_domains()  
    
    dv_dt = F/rho
    
    return compress(du_dt, dv_dt)
```

# IVP

```python
H = np.sqrt((Hy**2 + Hz**2)/8)
beta = (3*young + 2*l*(1-2*nu)**3 + 4*m*(1+nu)**2*(1-2*nu) 
        + 6*n*nu**2)/2/rho
c = np.sqrt(young/rho)
v = np.sqrt(1.2)*c
A = 3*(v**2 - c**2) / 2 / beta
    
def sln_dx(x, t, v, type='nr'):
    D = 1
    if type == 'nr':
        D = c/2/H*np.sqrt(3*(v**2 - c**2)/(2*(nu**2 + nu + 1)*c**2*v**2 
                                       - (nu + 1)*(c**4 + v**4)))
    if type == 'reg':
        D = 1/2/H/v/nu*np.sqrt(3*(v**2 - c**2)/2)
    return A * np.cosh(D*(x - v*t))**(-2)

def sln(x, t, v, type='nr'):
    D = 1
    if type == 'nr':
        D = c/2/H*np.sqrt(3*(v**2 - c**2)/(2*(nu**2 + nu + 1)*c**2*v**2 
                                       - (nu + 1)*(c**4 + v**4)))
    if type == 'reg':
        D = 1/2/H/v/nu*np.sqrt(3*(v**2 - c**2)/2)
    return A / D * np.tanh(D*(x - v*t))

x0 = Lx/2

X, Y, Z = np.meshgrid(*mesh.grid(), indexing='ij')
Ux = sln(X - x0, 0, v, type='reg')
Uy = -nu*Y*sln_dx(X - x0, 0, v, type='reg')
Uz = -nu*Z*sln_dx(X - x0, 0, v, type='reg')
u0 = TensorField(mesh, np.stack((Ux, Uy, Uz)))
v0 = -v*u0.diff(dim=0)
```

```python
def iv(x, ib, ob, A):
    if x <= -ob:
        return 0.0
    if x > -ib and x < ib:
        return A*(x + 2*(ob - ib)/np.pi * np.cos(np.pi/2*ob/(ob - ib)))
    if x > -ob and x <= -ib:
        return -2*A*(ob - ib)/np.pi * np.cos(np.pi/2*(x + ob)/(ob - ib))
    if x >= ib and x <= ob:
        return -A*np.sin(np.pi/2*(x - ob)/(ob - ib))

def ivf(x, ib, ob, A):
    if x < -ob:
        return 0.0, 0.0
    else:
        def func(x):
            return iv_x_arr(x, ib, ob, A)
        return scipy.integrate.quad(func, -ob, x)
    
def ivf_arr(x, ib, ob, A):
    if type(x) == float:
        x = np.asarray([x])
    res = np.zeros_like(x)
    def func(x):
        return iv_x(x, ib, ob, A)
    for i in range(x.shape[0]):
        res[i], a = scipy.integrate.quad(func, -ob, x[i])
    #res, a = scipy.integrate.quad(func, -ob, x)
    return res
    
def iv_x(x, ib, ob, A):
    if x <= -ob or x >= ob:
        return 0.0
    if x > -ob and x <= -ib:
        return A*np.sin(np.pi/2*(x + ob)/(ob - ib))
    if x > -ib and x < ib:
        return A
    if x >= ib and x <= ob:
        return -A*np.sin(np.pi/2*(x - ob)/(ob - ib))

def iv_x_arr(x, ib, ob, A):
    if type(x) == float:
        x = np.asarray([x])
    res = np.zeros_like(x)
    p = [(x[i] > -ob and x[i] <= -ib) for i in range(x.shape[0])]
    res[p] = A*np.sin(np.pi/2*(x[p] + ob)/(ob - ib))
    p = [(x[i] > -ib and x[i] < ib) for i in range(x.shape[0])]
    res[p] = A
    p = [(x[i] >= ib and x[i] <= ob) for i in range(x.shape[0])]
    res[p] = -A*np.sin(np.pi/2*(x[p] - ob)/(ob - ib))
    return res

def iv_x_new(x, ib, ob, A):
    def f1(x):
        return 0*x
    def f2(x):
        return A*np.sin(np.pi/2*(x + ob)/(ob - ib))
    def f3(x):
        return A + 0*x
    def f4(x):
        return -A*np.sin(np.pi/2*(x - ob)/(ob - ib))
    c1 = np.logical_or(x <= -ob, x >= ob)
    c2 = np.logical_and(x > -ob, x <= -ib)
    c3 = np.logical_and(x > -ib, x < ib)
    c4 = np.logical_and(x >= ib, x <= ob)
    return np.piecewise(x, [c1, c2, c3, c4], [f1, f2, f3, f4])

def ivf_new(x, ib, ob, A):
    def f1(x):
        return 0*x
    def f2(x):
        return -2*A*(ob - ib)/np.pi * (np.cos(np.pi/2*(x + ob)/(ob - ib)) - 1)
    def f3(x):
        return A*(x + ib) + f2(-ib)
    def f4(x):
        return 2*A*(ob - ib)/np.pi * np.cos(np.pi/2*(x - ob)/(ob - ib)) + f3(ib)
    def f5(x):
        return f4(ob)
    c1 = x <= -ob
    c2 = np.logical_and(x > -ob, x <= -ib)
    c3 = np.logical_and(x > -ib, x < ib)
    c4 = np.logical_and(x >= ib, x <= ob)
    c5 = x >= ob
    return np.piecewise(x, [c1, c2, c3, c4, c5], [f1, f2, f3, f4, f5])
```

```python
A = -0.01
L = 40
c = np.sqrt(young/rho)
v = np.sqrt(c**2 + 2/3*A)
x0 = Lx/2
ib = L/2
ob = L

X, Y, Z = np.meshgrid(*mesh.grid(), indexing='ij')
Ux = ivf_new(X - x0, ib, ob, A)
#Uy = -nu*A*Y*np.cosh((X - x0)/L)**-2
#Uz = -nu*A*Z*np.cosh((X - x0)/L)**-2
Uy = -nu*A*Y*iv_x_new(X - x0, ib, ob, A)
Uz = -nu*A*Z*iv_x_new(X - x0, ib, ob, A)
u0 = TensorField(mesh, np.stack((Ux, Uy, Uz)))
v0 = TensorField(mesh, np.zeros((3,) + mesh.shape))
#v0 = -c*u0.diff(dim=0)
```

```python
%%time
t_max = 10
dt = 1
T = np.arange(0, t_max + dt/2, dt)
U = TensorField(mesh, np.zeros((len(T), 3) + mesh.shape))
V = TensorField(mesh, np.zeros((len(T), 3) + mesh.shape))
E = np.zeros(len(T))

r = sp.integrate.ode(derivative).set_integrator('dop853', rtol=1e-10, atol=1e-10, nsteps=1e4)
r.set_initial_value(compress(u0, v0))
for k, t in enumerate(Progress(T)):
    if t > r.t:
        r.integrate(t)
    U[k], V[k] = decompress(r.y)
    E[k] = energy(U[k], V[k])
```
